```python
import matplotlib.pyplot as plt
import pickle
import os
```

## Reload Fraction Pickle Files

"Fractions" has 1000 elements which are two-item lists. The first item of the list represents the predicted fraction of dogs in the dog-cat test images; the second item of the list represents the predicted fraction of dogs in the other 8 classes test images.

```python
fraction_file = "resnet_balanced_256_1_batchnorm_true.pkl"
fractions_path = os.path.join("../results/fractions", fraction_file)
with open(fractions_path, "rb") as file:
    fractions = pickle.load(file)
```

## Plot

```python
y = [ls[0] for ls in fractions] # the 0 item represents the predicted fraction of dogs in the dog-cat test images
x = range(len(y))
plt.plot(x, y, label="256:1")
plt.legend(title="dog:cat")
```

```python
y = [ls[1] for ls in fractions] # the 0 item represents the predicted fraction of dogs in the dog-cat test images
x = range(len(y))
plt.plot(x, y, label="256:1")
plt.legend(title="dog:cat")
```
